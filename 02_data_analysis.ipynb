```python
# ======================================
# Zomato Bangalore Data Analysis & Visualization
# ======================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
sns.set(style='whitegrid')

# --------------------------------------
# 1 Load cleaned dataset
# --------------------------------------
# 1 Load cleaned dataset
df = pd.read_csv('../data/zomato_cleaned.csv')
print("Data loaded successfully! Shape:", df.shape)

# Standardize column names
df.rename(columns={'approx_cost(for two people)': 'approx_cost'}, inplace=True)

# --------------------------------------
# 2 Basic overview
# --------------------------------------
print("\n--- Basic Info ---")
print(df.info())
print("\n--- Sample Data ---")
display(df.head())

# --------------------------------------
# 3 Distribution of Restaurant Ratings
# --------------------------------------
plt.figure(figsize=(8,5))
sns.histplot(df['rate'], bins=20, kde=True)
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()

# --------------------------------------
# 4 Top 10 Most Popular Locations
# --------------------------------------
plt.figure(figsize=(10,5))
top_locations = df['location'].value_counts().head(10)
sns.barplot(x=top_locations.values, y=top_locations.index)
plt.title('Top 10 Most Popular Locations in Bangalore')
plt.xlabel('Number of Restaurants')
plt.ylabel('Location')
plt.show()

# --------------------------------------
# 5 Average Rating by Restaurant Type
# --------------------------------------
plt.figure(figsize=(10,5))
type_rating = df.groupby('rest_type')['rate'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=type_rating.values, y=type_rating.index)
plt.title('Top 10 Restaurant Types by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Restaurant Type')
plt.show()

# --------------------------------------
# 6 Correlation between Cost and Rating
# --------------------------------------
plt.figure(figsize=(6,5))
sns.scatterplot(data=df, x='approx_cost', y='rate', alpha=0.5)
plt.title('Correlation between Approx Cost (for two) and Rating')
plt.xlabel('Approx Cost (for two)')
plt.ylabel('Rating')
plt.show()

corr = df['approx_cost'].corr(df['rate'])
print(f"\n Correlation between cost and rating: {corr:.2f}")

# --------------------------------------
# 7 Top Cuisines in Bangalore
# --------------------------------------
plt.figure(figsize=(10,5))
top_cuisines = df['cuisines'].value_counts().head(10)
sns.barplot(x=top_cuisines.values, y=top_cuisines.index)
plt.title('Top 10 Most Common Cuisines in Bangalore')
plt.xlabel('Number of Restaurants')
plt.ylabel('Cuisines')
plt.show()

# Scatter plot
plt.figure(figsize=(6,5))
sns.scatterplot(data=df, x='approx_cost', y='rate', alpha=0.5)
plt.title('Correlation between Approx Cost (for two) and Rating')
plt.xlabel('Approx Cost (for two)')
plt.ylabel('Rating')
plt.show()

# --------------------------------------
# 8 Key Insights Summary
# --------------------------------------
print("\n Key Insights:")
print("1. The highest number of restaurants are concentrated in top 3 localities:", list(top_locations.index[:3]))
print("2. Higher average ratings are found in fine dining and casual dining restaurants.")
print(f"3. Correlation between cost and rating is {corr:.2f} (moderate relationship).")
print("4. North Indian and Chinese cuisines dominate the Bangalore dining scene.")

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

# Make charts look cleaner
plt.rcParams['figure.figsize'] = (8, 5)

# ------------------------------------------------
# 1 Rating Distribution
# ------------------------------------------------
if 'rate' in df.columns:
    plt.figure()
    sns.histplot(df['rate'], bins=20, kde=True, color='skyblue')
    plt.title('Rating Distribution')
    plt.xlabel('Average Rating')
    plt.ylabel('Count')
    plt.show()

# ------------------------------------------------
# 2 Top 10 Locations by Restaurant Count
# ------------------------------------------------
if 'location' in df.columns:
    plt.figure()
    top_locations = df['location'].value_counts().nlargest(10)
    sns.barplot(x=top_locations.values, y=top_locations.index, palette='mako')
    plt.title('Top 10 Locations by Number of Restaurants')
    plt.xlabel('Number of Restaurants')
    plt.ylabel('Location')
    plt.show()

# ------------------------------------------------
# 3 Top 10 Cuisines
# ------------------------------------------------
if 'cuisines' in df.columns:
    plt.figure()
    top_cuisines = df['cuisines'].value_counts().nlargest(10)
    sns.barplot(x=top_cuisines.values, y=top_cuisines.index, palette='viridis')
    plt.title('Top 10 Cuisines in Bangalore')
    plt.xlabel('Number of Restaurants')
    plt.ylabel('Cuisine Type')
    plt.show()

# ------------------------------------------------
# 4 Online Delivery vs Ratings
# ------------------------------------------------
if 'online_order' in df.columns:
    plt.figure()
    sns.boxplot(x='online_order', y='rate', data=df, palette='Set2')
    plt.title('Online Order Availability vs Ratings')
    plt.xlabel('Online Order Available')
    plt.ylabel('Rating')
    plt.show()

# ------------------------------------------------
# 5 Approx Cost vs Rating (correlation plot)
# ------------------------------------------------
if 'approx_cost' in df.columns and 'rate' in df.columns:
    plt.figure()
    sns.scatterplot(data=df, x='approx_cost', y='rate', alpha=0.5, color='coral')
    plt.title('Approx Cost (for two) vs Rating')
    plt.xlabel('Approx Cost (for two)')
    plt.ylabel('Rating')
    plt.show()

# ------------------------------------------------
# 6 Table Booking vs Rating
# ------------------------------------------------
if 'book_table' in df.columns:
    plt.figure()
    sns.boxplot(x='book_table', y='rate', data=df, palette='Set1')
    plt.title('Table Booking Availability vs Rating')
    plt.xlabel('Table Booking Available')
    plt.ylabel('Rating')
    plt.show()

```

```python
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Image
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib.pagesizes import A4
from reportlab.lib.units import inch
import matplotlib.pyplot as plt
import seaborn as sns
import os

# -------------------------------------------
# 1 Setup
# -------------------------------------------
report_path = "../Zomato_Bangalore_Analysis_Report.pdf"
doc = SimpleDocTemplate(report_path, pagesize=A4)
styles = getSampleStyleSheet()
story = []

story.append(Paragraph("<b>Zomato Bangalore Analysis  Business Insights Report</b>", styles['Title']))
story.append(Spacer(1, 12))

# -------------------------------------------
# 2 Summary Section
# -------------------------------------------
story.append(Paragraph("<b>Dataset Summary</b>", styles['Heading2']))
story.append(Paragraph(f"Total Records: {len(df)}", styles['Normal']))

if 'location' in df.columns:
    story.append(Paragraph(f"Unique Locations: {df['location'].nunique()}", styles['Normal']))
if 'cuisines' in df.columns:
    story.append(Paragraph(f"Unique Cuisines: {df['cuisines'].nunique()}", styles['Normal']))
if 'rate' in df.columns:
    story.append(Paragraph(f"Average Rating: {df['rate'].mean():.2f}", styles['Normal']))
if 'approx_cost' in df.columns:
    story.append(Paragraph(f"Average Cost for Two: {df['approx_cost'].mean():.0f}", styles['Normal']))

story.append(Spacer(1, 12))

# -------------------------------------------
# 3 Save charts temporarily & add to report
# -------------------------------------------
img_dir = "charts"
os.makedirs(img_dir, exist_ok=True)

charts = []

def save_chart(fig_name, plot_func):
    """Helper to generate and save each chart."""
    path = os.path.join(img_dir, fig_name)
    plot_func()
    plt.tight_layout()
    plt.savefig(path)
    plt.close()
    charts.append(path)

# Generate charts
if 'rate' in df.columns:
    save_chart("rating_distribution.png", lambda: sns.histplot(df['rate'], bins=20, kde=True, color='skyblue'))
if 'location' in df.columns:
    save_chart("top_locations.png", lambda: sns.countplot(y=df['location'], order=df['location'].value_counts().head(10).index))
if 'cuisines' in df.columns:
    save_chart("top_cuisines.png", lambda: sns.countplot(y=df['cuisines'], order=df['cuisines'].value_counts().head(10).index))

# Add charts to PDF
story.append(Paragraph("<b>Visual Insights</b>", styles['Heading2']))
for img in charts:
    story.append(Image(img, width=5.5*inch, height=3.5*inch))
    story.append(Spacer(1, 12))

# -------------------------------------------
# 4 Save PDF
# -------------------------------------------
doc.build(story)
print(f" Business report successfully generated at: {report_path}")
```
