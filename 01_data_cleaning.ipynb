```python
import pandas as pd

# Load dataset
df = pd.read_csv('data/zomato.csv')
print("Original shape:", df.shape)

# Drop unnecessary columns
df.drop(columns=['url', 'address', 'phone'], inplace=True, errors='ignore')

# Clean 'rate' column (remove '/5', convert to float)
df['rate'] = df['rate'].astype(str).apply(lambda x: x.replace('/5', '').strip() if '/5' in x else x)
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')

# Clean 'approx_cost(for two people)'
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(str).replace(',', '', regex=True).astype(float)

# Fill missing values
before = df.isnull().sum().sum()
df.fillna({
    'rate': df['rate'].mean(),
    'approx_cost(for two people)': df['approx_cost(for two people)'].mean(),
    'location': df['location'].mode()[0],
    'cuisines': df['cuisines'].mode()[0]
}, inplace=True)
after = df.isnull().sum().sum()

print(f"✅ Missing values reduced by {((before - after)/before)*100:.2f}%")

# Save cleaned data
df.to_csv('data/zomato_cleaned.csv', index=False)
print("✅ Cleaned dataset saved as zomato_cleaned.csv")
```
